```python
# To display (you need ipyleaflet installed)
# !pip install ipyleaflet

import numpy as np
import rasterio
import matplotlib.pyplot as plt

from ipyleaflet import Map, ImageOverlay, LayersControl

with rasterio.open('dem.tif') as src:
    dem = src.read(1, masked=True)
    tif_bounds = src.bounds  # left, bottom, right, top

arr = dem.filled(np.nan)

vmin = np.nanpercentile(arr, 2)
vmax = np.nanpercentile(arr, 98)
scaled = np.clip((arr - vmin) / (vmax - vmin), 0, 1)

png_path = 'dem_overlay.png'
plt.imsave(png_path, scaled, origin='upper', cmap='terrain')

m = Map(
    center=((tif_bounds.bottom + tif_bounds.top) / 2, (tif_bounds.left + tif_bounds.right) / 2),
    zoom=8,
)

overlay = ImageOverlay(
    url=png_path,
    bounds=((tif_bounds.bottom, tif_bounds.left), (tif_bounds.top, tif_bounds.right)),
    name='DEM',
)

m.add(overlay)
m.add(LayersControl())
m
```
